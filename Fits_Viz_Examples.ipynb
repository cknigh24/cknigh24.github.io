#### Output quick visualizations of Monoceros R2 in all four WISE Bands 

```python
#Import WISE images from Shared Google Drive
# Make take a minute, each file is 65 mb

# Use function that extracts files from Drive to output location
from drive import drive_import as di

# output location
fitspath = 'C://Users//Collin//Desktop//Git_Project//monR2//fits//'

# use Drive id for specific files
file_id = '1QocubsGtx2L4a1GjAD0zRVydMI_IQeGL'
destination = fitspath + 'monR2_w1.fits'
di(file_id, destination)

file_id = '1UoLGjOtsqN_lFI3F5geeBuRmlDGf7yrk'
destination = fitspath + 'monR2_w2.fits'
di(file_id, destination)

file_id = '1J9NjpzAxFhsvOs9BhAUmrnHNSnOXE7tL'
destination = fitspath + 'monR2_w3.fits'
di(file_id, destination)

file_id = '1DA_wtTH9BYBAfBf-ELTiAxRcmyDivnK7'
destination = fitspath + 'monR2_w4.fits'
di(file_id, destination)
```

```python
# Import function to output quick visualizations with fits input
from fits_viz_template import fits_viz as viz
```

```python
# Set default parameters based on the results of MonR2_Intro Notebook
a1 = 600
a2 = 2400
b1 = 1600
b2 = 4095
pixel_size = 1.375
pcent = [0.25, 98.5]
plotpath = 'C://Users//Collin//Desktop//Git_Project//monR2//images//'
```

```python
# First run on WISE 1 (convertion factor 1.9350e-6 Jy/DN)
viz(filename = fitspath+'monr2_w1.fits',a1= a1,a2 = a2, b1 = b1, b2 = b2, 
    lower_p= pcent[0] ,upper_p=pcent[1], scaling = "sqrt", hist_display = True, 
    cumhist_display=True, fitspath=fitspath, plotpath = plotpath,
    save = True, wcs_axes=True, print_hdr= True,
    convert_factor = 1.9350e-6 ,pixel_size = pixel_size)
```

```python
# Run on WISE 2 (convertion factor 2.7048e-6 Jy/DN)
viz(filename = fitspath+'monr2_w2.fits',a1= a1,a2 = a2, b1 = b1, b2 = b2, 
    lower_p= pcent[0] ,upper_p=pcent[1], scaling = "sqrt", hist_display = True, 
    cumhist_display=True, fitspath=fitspath, plotpath = plotpath,
    save = True, wcs_axes=True, print_hdr= True,
    convert_factor = 2.7048e-6 ,pixel_size = pixel_size)
```

```python
# Run on WISE 3 (convertion factor 2.9045e-6 Jy/DN)
viz(filename = fitspath+'monr2_w3.fits',a1= a1,a2 = a2, b1 = b1, b2 = b2, 
    lower_p= pcent[0] ,upper_p=pcent[1]+1, scaling = "sqrt", hist_display = True, 
    cumhist_display=True, fitspath=fitspath, plotpath = plotpath,
    save = True, wcs_axes=True, print_hdr= True,
    convert_factor = 2.9045e-6 ,pixel_size = pixel_size)
```

```python
# Run on WISE 4 (convertion factor 5.2269e-5 Jy/DN)
viz(filename = fitspath+'monr2_w4.fits',a1= a1,a2 = a2, b1 = b1, b2 = b2, 
    lower_p= pcent[0] ,upper_p=pcent[1]+1, scaling = "sqrt", hist_display = True, 
    cumhist_display=True, fitspath=fitspath, plotpath = plotpath,
    save = True, wcs_axes=True, print_hdr= True,
    convert_factor = 5.2269e-5 ,pixel_size = pixel_size)
```
